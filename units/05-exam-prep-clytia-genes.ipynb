```python
import pandas as pd
```

```python
tcons_to_gene = pd.read_csv("/Users/npapadop/Documents/teaching/2023/advanced_scRNAseq/data/clytia/xloc_pairs.csv",
                            index_col=1, header=None, sep=" ")
```

```python
infile = "/Users/npapadop/Documents/teaching/2023/advanced_scRNAseq/data/clytia/transcripts_TRINITY_20201021.fa"
outfile = "/Users/npapadop/Documents/teaching/2023/advanced_scRNAseq/data/clytia/transcripts_xloc.fa"
with open(outfile, "w") as out:
    with open(infile) as f:
        for line in f:
            if line.startswith(">"):
                tcons = line.strip()[1:]
                gene = tcons_to_gene.loc[tcons].values[0]
                out.write(">" + tcons + "|" + gene + "\n")
            else:
                out.write(line)
```

```python
def keep_higher_score_duplicate(df, drop_col=1):
    df = df.sort_values(11, ascending=False)
    return df.drop_duplicates(drop_col)
```

```python
ntc = pd.read_csv("~/Documents/teaching/2023/advanced_scRNAseq/data/exam/aln/nvec_to_chem.m8.raw", sep="\t", header=None)
ntc[1] = ntc[1].str.split("|").str[1]
ntc.groupby(0).apply(keep_higher_score_duplicate).to_csv("~/Documents/teaching/2023/advanced_scRNAseq/data/exam/aln/nvec_to_chem.m8", sep="\t", header=None, index=None)
```

```python
ctn = pd.read_csv("~/Documents/teaching/2023/advanced_scRNAseq/data/exam/aln/chem_to_nvec.m8.raw", sep="\t", header=None)
ctn[0] = ctn[0].str.split("|").str[1]
ctn.groupby(0).apply(keep_higher_score_duplicate).to_csv("~/Documents/teaching/2023/advanced_scRNAseq/data/exam/aln/chem_to_nvec.m8", sep="\t", header=None, index=None)
```
